Start by importing the modules we need.

```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

First, let's import our data and split it into training and test data.

```python
df = pd.read_csv('data.csv')
y = df['Stress Level'].values
X = df.drop('Stress Level', axis=1).values
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=5, stratify=y)
```

Next, we'll determine the best K value to use.

```python
scores = []
k_values = []

for k in range(1, 21):
    k_values.append(k)

    knn = KNeighborsClassifier(n_neighbors=k)
    # Train the model
    knn.fit(X_train, y_train)
    # Compute score
    scores.append(knn.score(X_test, y_test))

plt.plot(k_values, scores)
plt.xticks(k_values)
plt.show()
```

We have plenty of good *k* options here, but I'll use 15.

```python
knn = KNeighborsClassifier(n_neighbors=15)
knn.fit(X_train, y_train)
prediction = knn.predict(X_test)
print(prediction)
```
